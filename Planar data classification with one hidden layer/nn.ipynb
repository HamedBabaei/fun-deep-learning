## Import modules

```python
from nn import NN
import util
import __init__
from utils.dataloader import load_dataset
import matplotlib.pyplot as plt
```

## Loading Data

```python
X, Y = load_dataset(None, 'planer')

# Visualize the data: my changes was to put Y[0] in action
util.plot_scatter(X,Y)
```

##  Simple Logistic Regression

```python
# Train the logistic regression classifier
import sklearn.linear_model
clf = sklearn.linear_model.LogisticRegressionCV()
clf.fit(X.T, Y[0].T)
```

```python
# Plot the decision boundary for logistic regression
print("Logistic Regression")
util.plot_decision_boundary(lambda x: clf.predict(x), X, Y[0])
```

```python
import numpy as np
# Print accuracy
LR_predictions = clf.predict(X.T)
print ('Accuracy of logistic regression: %d ' % 
       float((np.dot(Y,LR_predictions) + np.dot(1-Y,1-LR_predictions))/float(Y.size)*100) +
       '% ' + "(percentage of correctly labelled datapoints)")    
```

## Neural Network Classification

```python
model = NN(hidden_layer_size=50, num_iterations=10000, learning_rate=0.4 )

model.fit(X, Y, verbose=True)
```

```python
preds = model.predict(X)
print ('Accuracy: %d' % float((np.dot(Y,preds.T) + np.dot(1-Y,1-preds.T))/float(Y.size)*100) + '%')
```
