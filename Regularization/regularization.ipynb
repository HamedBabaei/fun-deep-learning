```python
import utils
import numpy as np
```

```python
X_train, Y_train, X_test, Y_test = utils.load_2D_dataset(path='../dataset/2D-dataset/data.mat', plot=True)
```

```python
from dnn import LLayerModel

layers_dims = [X_train.shape[0], 20, 3, 1]

model = LLayerModel(layers_dims, num_iterations = 3000, learning_rate = 0.3)
```

```python
model.fit(X_train, Y_train, verbose=False, plot_cost=True)
```

```python
predicts = model.predict(X_train)

print("ACC of Train:", np.sum((predicts == Y_train)/X_train.shape[1]))

predicts = model.predict(X_test)

print("ACC of Test:", np.sum((predicts == Y_test)/X_test.shape[1]))
```
