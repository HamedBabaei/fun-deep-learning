## 1. Loading Dataset

```python
import utils
import numpy as np

X_train, Y_train, X_test, Y_test = utils.load_2D_dataset(path='../dataset/2D-dataset/data.mat', plot=True)
```

## 2. L-Layer Model Without Regularization

```python
from dnn import LLayerModel

layers_dims = [X_train.shape[0], 20, 3, 1]

model = LLayerModel(layers_dims, num_iterations = 30000, learning_rate = 0.03)

model.fit(X_train, Y_train, verbose=False, plot_cost=True)
```

```python
predicts = model.predict(X_train)

print("ACC of Train:", np.mean((predicts == Y_train)))

predicts = model.predict(X_test)

print("ACC of Test:", np.mean((predicts == Y_test)))
```

```python
func = lambda x: model.predict_dec(x.T)
utils.plot_decision_boundary(func, X_train, Y_train[0])
```

## 3. L-Layer Model With Regularization

```python
from dnn import LLayerModel

layers_dims = [X_train.shape[0], 20, 3, 1]

model = LLayerModel(layers_dims, num_iterations = 30000, learning_rate = 0.03, lambd = 0.7)
model.fit(X_train, Y_train, verbose=False, plot_cost=True)

predicts = model.predict(X_train)
print("ACC of Train:", np.mean((predicts == Y_train)))
predicts = model.predict(X_test)
print("ACC of Test:", np.mean((predicts == Y_test)))


func = lambda x: model.predict_dec(x.T)
utils.plot_decision_boundary(func, X_train, Y_train[0])
```
