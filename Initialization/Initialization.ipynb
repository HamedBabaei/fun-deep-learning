## 1. Loading Dataset

```python
import utils
import numpy as np

X_train, Y_train, X_test, Y_test = utils.load_dataset()
```

## 2 - deep initialization

```python
from dnn import LLayerModel

layers_dims = [X_train.shape[0], 10, 5, 1]

model = LLayerModel(layers_dims, num_iterations = 15000, learning_rate = 0.01, initialization='none' )
model.fit(X_train, Y_train, verbose=False, plot_cost=True)

predicts = model.predict(X_train)
print("ACC of Train:", np.mean((predicts == Y_train)))
predicts = model.predict(X_test)
print("ACC of Test:", np.mean((predicts == Y_test)))

func = lambda x: model.predict_dec(x.T)
utils.plot_decision_boundary(func, X_train, Y_train[0])
```
