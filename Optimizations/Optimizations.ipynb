## 1. Loading Dataset

```python
import utils
import numpy as np

X_train, Y_train = utils.load_dataset()
```

## 2 - Batch-Gradient descent

```python
from dnn import LLayerModel

layers_dims = [X_train.shape[0], 5, 2, 1]

model = LLayerModel(layers_dims, num_iterations = 10000, learning_rate = 0.0007, initialization='none',
                    mini_batch_size = 1, optimizer='gd')
model.fit(X_train, Y_train, verbose=True, plot_cost=True)
    
predicts = model.predict(X_train)
print("ACC of Train:", np.mean((predicts == Y_train)))

func = lambda x: model.predict_dec(x.T)
utils.plot_decision_boundary(func, X_train, Y_train[0])
```
